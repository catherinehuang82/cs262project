```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv('logs/game0/game_log_0.csv', index_col=0)

df0
```

```python
plt.scatter(df0.index, df0['total_payout'], s=1)
plt.xlabel('Encounter Number')
plt.ylabel('Total Payout')
plt.title('Payout Over Time of One Run, 5000 Encounters')
```

```python
plt.scatter(df0.index[:1000], df0['total_payout'][:1000], s=1)
plt.xlabel('Encounter Number')
plt.ylabel('Total Payout')
plt.title('Payout Over Time of One Run, 1000 Encounters')
```
