```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# total agent payout over time
## payouts = (2, -2)

```python
df0 = pd.read_csv('logs/game0/game_log_0.csv', index_col=0)

df0
```

```python
plt.scatter(df0.index[:3000], df0['total_payout'][:3000], s=1)
plt.xlabel('Encounter Number')
plt.ylabel('Total Payout')
plt.title('Payout Over Time of One Run, 3000 Encounters')
```

```python
plt.scatter(df0.index[:1000], df0['total_payout'][:1000], s=1)
plt.xlabel('Encounter Number')
plt.ylabel('Total Payout')
plt.title('Payout Over Time of One Run, 1000 Encounters')
```

```python
plt.scatter(df0.index[:100], df0['total_payout'][:100], s=1)
plt.xlabel('Encounter Number')
plt.ylabel('Total Payout')
plt.title('Payout Over Time of One Run, 100 Encounters')
```
